### üî¢Q5. [5 points] ‚Äî Compare Models of Different Complexity

Run the following code to compare bias and variance visually by fitting linear and polynomial models to noisy data.


```python
import matplotlib.pyplot as plt

# Simulated model complexity and corresponding errors
model_complexity = [1, 2, 3, 4, 5, 6, 7, 8]
training_error = [8, 6, 4, 3, 2, 1.5, 1.2, 1]
validation_error = [9, 7, 5, 4, 4.5, 5.5, 7, 9]

# Plotting the errors
plt.figure(figsize=(8, 4))
plt.plot(model_complexity, training_error, label="Training Error", marker='o')
plt.plot(model_complexity, validation_error, label="Validation Error", marker='o')

# Adding vertical lines for regions
plt.axvline(x=2.5, color='gray', linestyle='--', label='Low/Med Boundary')
plt.axvline(x=5.5, color='gray', linestyle='--', label='Med/High Boundary')

# Labels and title
plt.xlabel("Model Complexity")
plt.ylabel("Error")
# plt.title("Bias‚ÄìVariance Trade-off (with Complexity Regions)")
plt.legend()
plt.grid(True)

# Text labels for regions
plt.text(1, 9.5, "Low\nComplexity", ha='center', fontsize=10, color='gray')
plt.text(4, 9.5, "Medium\nComplexity", ha='center', fontsize=10, color='gray')
plt.text(7, 9.5, "High\nComplexity", ha='center', fontsize=10, color='gray')

plt.show()
```
