### ðŸ’» Q5. Code Completion: Estimate Bias and Variance from Simulation (5 points)

You are given code that simulates repeated training of a regression model to estimate prediction **bias and variance** at a fixed input value `x0 = 1.5`. Some parts of the code are missing â€” complete them to compute:

- The **mean prediction** at `x0`
- The **bias** using the known true function value
- The **variance** of the predictions


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample

np.random.seed(42)
X_full = np.linspace(0, 1, 100).reshape(-1, 1)
y_true = np.sin(2 * np.pi * X_full).ravel()

plt.figure(figsize=(8, 4))
for _ in range(15):
    # Simulate random training data
    X = np.linspace(0, 1, 20).reshape(-1, 1)
    y = np.sin(2 * np.pi * X).ravel() + np.random.normal(0, 0.3, size=20)

    # Bootstrap sample
    X_sample, y_sample = resample(X, y)
    model = LinearRegression()
    model.fit(X_sample, y_sample)

    # Predict and plot
    y_pred = model.predict(X_full)
    plt.plot(X_full, y_pred, color='red', alpha=0.3)

# Plot true function
plt.plot(X_full, y_true, 'k--', label='True Function')
plt.title("Model Predictions vs True Function")
plt.legend()
plt.show()
```

1. Does the average of red curves follow the true function well?
If yes, the bias is low, else high.

2. Do the red curves vary a lot from one another?
If yes, the variance is high, else low.

3. Based on your observations, does this model have high bias, high variance, both, or neither?
